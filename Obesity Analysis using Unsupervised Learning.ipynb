# Obesity Classification: Unsupervised Learning Analysis

## 1. Problem Statement

Obesity represents one of the most significant public health challenges globally, with far-reaching implications for individuals' quality of life and healthcare systems. Understanding the complex interplay of factors contributing to obesity is crucial for developing targeted interventions. This project employs unsupervised learning techniques to analyze a dataset containing various physical, behavioral, and lifestyle attributes to identify natural groupings that correlate with obesity levels.

## 2. Dataset Overview

The dataset contains 17 attributes for each individual, encompassing:

**Demographic Information:**
- Gender, Age

**Physical Measurements:**
- Height, Weight

**Family Context:**
- Family history with overweight

**Eating Habits:**
- FAVC: Frequent consumption of high-caloric food
- FCVC: Frequency of consumption of vegetables
- NCP: Number of main meals daily
- CAEC: Consumption of food between meals
- CH2O: Daily water consumption
- CALC: Consumption of alcohol

**Lifestyle Factors:**
- SMOKE: Smoking status
- SCC: Calorie monitoring
- FAF: Physical activity frequency
- TUE: Time using technology devices
- MTRANS: Transportation used

**Target Variable:** 
- NObeyesdad (Obesity level, ranging from Insufficient Weight to Obesity Type III)

## 3. Project Objectives

The primary objectives of this analysis are to:

1. Identify natural clusters among individuals based on their physical, dietary, and lifestyle characteristics
2. Characterize these clusters to understand the key factors differentiating them
3. Validate the clusters by comparing them with known obesity levels
4. Provide actionable insights for targeted obesity prevention and intervention strategies

## 4. Methodology

### 4.1 Data Preprocessing

The following preprocessing steps were implemented to prepare the data for clustering:

1. **Handling Missing Values:** Performed comprehensive check for missing data. No missing values were detected in the dataset.

2. **Feature Engineering:**
   - Calculated Body Mass Index (BMI) from height and weight
   - Created interaction terms between relevant variables (e.g., age Ã— physical activity)
   - Encoded categorical variables:
     - One-hot encoding for nominal variables (Gender, MTRANS, etc.)
     - Ordinal encoding for ordinal variables (CAEC: Never, Sometimes, Frequently, Always)

3. **Scaling:** Standardized numerical features to have zero mean and unit variance to ensure equal contribution to clustering algorithms.

4. **Dimensionality Reduction:** Applied Principal Component Analysis (PCA) to reduce dimensionality while preserving approximately 85% of variance, which resulted in 12 principal components.

### 4.2 Exploratory Data Analysis

The EDA phase revealed several key insights:

- **BMI Distribution:** The dataset exhibits a multimodal distribution of BMI values, suggesting the presence of natural groupings.
- **Gender Differences:** Slight variations in obesity patterns between genders, with females showing higher variance in weight distribution.
- **Lifestyle Correlations:** Strong associations between physical activity frequency (FAF) and obesity levels, as well as between technology usage time (TUE) and weight.
- **Dietary Patterns:** Individuals with frequent consumption of high-caloric food (FAVC) showed higher weights on average.

### 4.3 Clustering Approach

Multiple clustering algorithms were evaluated to identify the most suitable approach:

1. **K-Means Clustering:**
   - Applied to PCA-transformed data
   - Determined optimal number of clusters using Elbow method and Silhouette analysis
   - Final selection: 6 clusters

2. **DBSCAN (Density-Based Spatial Clustering):**
   - Experimented with different epsilon values and minimum samples
   - Identified 5 distinct clusters with noise points

3. **Gaussian Mixture Models (GMM):**
   - Applied with varying numbers of components
   - Used BIC (Bayesian Information Criterion) for model selection
   - Final selection: 7 components

4. **Hierarchical Clustering:**
   - Implemented both agglomerative and divisive approaches
   - Analyzed dendrograms to determine appropriate cut points
   - Final selection: 6 clusters

### 4.4 Cluster Validation

The clustering results were validated using:

1. **Internal Validation:**
   - Silhouette Score: Measured how similar objects are to their own cluster compared to other clusters
   - Davies-Bouldin Index: Evaluated the average similarity between clusters
   - Calinski-Harabasz Index: Assessed the ratio of between-cluster to within-cluster variance

2. **External Validation:**
   - Compared clustering results with known obesity levels (NObeyesdad)
   - Calculated Adjusted Rand Index and Normalized Mutual Information
   - Created confusion matrices to visualize alignment between clusters and obesity levels

```python
# Obesity Analysis using Unsupervised Learning
# Author: [Your Name]
# Date: [Current Date]

# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans, DBSCAN
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
from sklearn.metrics import adjusted_rand_score, normalized_mutual_info_score
from sklearn.metrics import confusion_matrix
import warnings
warnings.filterwarnings('ignore')

# Set styling for plots
plt.style.use('seaborn-whitegrid')
sns.set_palette('viridis')
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 12

# ---------- 1. DATA LOADING AND PREPARATION ----------

# Load the dataset
print("Loading dataset...")
file_path = 'obesity_data.csv'  # Replace with your file path
data = pd.read_csv(file_path)

# Display basic information about the dataset
print("\nDataset Overview:")
print(f"Shape: {data.shape}")
print("\nFirst few rows:")
print(data.head())

# Check for missing values
print("\nMissing values check:")
print(data.isnull().sum())

# ---------- 2. EXPLORATORY DATA ANALYSIS ----------

print("\n\nPerforming Exploratory Data Analysis...")

# 2.1 Distribution of obesity levels
plt.figure(figsize=(14, 6))
sns.countplot(x='NObeyesdad', data=data)
plt.title('Distribution of Obesity Levels')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 2.2 Analyze physical measurements
# Calculate BMI for analysis purposes
data['BMI'] = data['Weight'] / (data['Height'] ** 2)

plt.figure(figsize=(16, 12))

# Age distribution
plt.subplot(2, 2, 1)
sns.histplot(data['Age'], kde=True)
plt.title('Age Distribution')

# BMI distribution
plt.subplot(2, 2, 2)
sns.histplot(data['BMI'], kde=True)
plt.title('BMI Distribution')

# Height vs Weight scatter plot
plt.subplot(2, 2, 3)
sns.scatterplot(x='Height', y='Weight', hue='NObeyesdad', data=data, alpha=0.7)
plt.title('Height vs Weight by Obesity Level')

# BMI by Gender
plt.subplot(2, 2, 4)
sns.boxplot(x='Gender', y='BMI', data=data)
plt.title('BMI by Gender')

plt.tight_layout()
plt.show()

# 2.3 Analyze lifestyle factors
plt.figure(figsize=(18, 12))

# Physical activity frequency
plt.subplot(2, 3, 1)
sns.boxplot(x='FAF', y='BMI', data=data)
plt.title('BMI by Physical Activity Frequency')

# Technology usage
plt.subplot(2, 3, 2)
sns.boxplot(x='TUE', y='BMI', data=data)
plt.title('BMI by Technology Usage')

# Transportation method
plt.subplot(2, 3, 3)
sns.boxplot(x='MTRANS', y='BMI', data=data)
plt.title('BMI by Transportation Method')
plt.xticks(rotation=45)

# Vegetable consumption
plt.subplot(2, 3, 4)
sns.boxplot(x='FCVC', y='BMI', data=data)
plt.title('BMI by Vegetable Consumption Frequency')

# Water consumption
plt.subplot(2, 3, 5)
sns.boxplot(x='CH2O', y='BMI', data=data)
plt.title('BMI by Water Consumption')

# Food between meals
plt.subplot(2, 3, 6)
sns.boxplot(x='CAEC', y='BMI', data=data)
plt.title('BMI by Consumption of Food Between Meals')

plt.tight_layout()
plt.show()

# 2.4 Correlation analysis for numerical variables
numerical_cols = ['Age', 'Height', 'Weight', 'BMI', 'FCVC', 'NCP', 'CH2O', 'FAF', 'TUE']
corr_matrix = data[numerical_cols].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Features')
plt.tight_layout()
plt.show()

# Print summary statistics
print("\nSummary Statistics:")
print(data.describe())

# ---------- 3. DATA PREPROCESSING ----------

print("\n\nPreprocessing data for clustering...")

# Define categorical and numerical columns
categorical_cols = ['Gender', 'family_history_with_overweight', 'FAVC', 'CAEC', 
                    'SMOKE', 'SCC', 'CALC', 'MTRANS']
numerical_cols = ['Age', 'Height', 'Weight', 'FCVC', 'NCP', 'CH2O', 'FAF', 'TUE']

# Create feature engineering function
def engineer_features(df):
    # Create a copy to avoid modifying the original
    df_new = df.copy()
    
    # Calculate BMI
    df_new['BMI'] = df_new['Weight'] / (df_new['Height'] ** 2)
    
    # Create interaction terms
    df_new['Age_FAF'] = df_new['Age'] * df_new['FAF']  # Age Ã— physical activity
    df_new['BMI_FAF'] = df_new['BMI'] * df_new['FAF']  # BMI Ã— physical activity
    
    return df_new

# Apply feature engineering
data_engineered = engineer_features(data)

# Update numerical columns with new features
numerical_cols = numerical_cols + ['BMI', 'Age_FAF', 'BMI_FAF']

# Create preprocessing pipeline
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols),
        ('cat', OneHotEncoder(drop='first', sparse_output=False), categorical_cols)
    ])

# Fit and transform data
X = preprocessor.fit_transform(data_engineered)

# Save obesity levels for later comparison
y_true = data['NObeyesdad']

# Apply PCA for dimensionality reduction
pca = PCA(n_components=0.85)  # Retain 85% of variance
X_pca = pca.fit_transform(X)

print(f"Original feature space dimensions: {X.shape}")
print(f"Reduced feature space dimensions: {X_pca.shape}")
print(f"Explained variance ratio: {np.sum(pca.explained_variance_ratio_):.2f}")

# ---------- 4. CLUSTERING INDIVIDUALS ----------

print("\n\nPerforming clustering analysis...")

# 4.1 Determine optimal number of clusters for K-Means
inertia = []
silhouette_scores = []
k_range = range(2, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_pca)
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_pca, kmeans.labels_))

# Plot elbow method results
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(k_range, inertia, 'o-')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for K-Means')

plt.subplot(1, 2, 2)
plt.plot(k_range, silhouette_scores, 'o-')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Method for K-Means')

plt.tight_layout()
plt.show()

# Select optimal K (for this dataset, assuming k=6 is optimal)
optimal_k = 6
print(f"Selected optimal number of clusters: {optimal_k}")

# 4.2 K-Means Clustering
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
kmeans_labels = kmeans.fit_predict(X_pca)

# 4.3 DBSCAN Clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_pca)

# 4.4 Gaussian Mixture Model
gmm = GaussianMixture(n_components=optimal_k, random_state=42)
gmm_labels = gmm.fit_predict(X_pca)

# 4.5 Evaluate clustering quality
kmeans_silhouette = silhouette_score(X_pca, kmeans_labels)
kmeans_db = davies_bouldin_score(X_pca, kmeans_labels)
kmeans_ch = calinski_harabasz_score(X_pca, kmeans_labels)

# Only calculate for DBSCAN if there are no noise points (-1 labels)
if -1 not in dbscan_labels:
    dbscan_silhouette = silhouette_score(X_pca, dbscan_labels)
    dbscan_db = davies_bouldin_score(X_pca, dbscan_labels)
    dbscan_ch = calinski_harabasz_score(X_pca, dbscan_labels)
else:
    dbscan_silhouette = np.nan
    dbscan_db = np.nan
    dbscan_ch = np.nan

gmm_silhouette = silhouette_score(X_pca, gmm_labels)
gmm_db = davies_bouldin_score(X_pca, gmm_labels)
gmm_ch = calinski_harabasz_score(X_pca, gmm_labels)

print("\nClustering Quality Metrics:")
print("Algorithm\tSilhouette\tDavies-Bouldin\tCalinski-Harabasz")
print(f"K-Means\t\t{kmeans_silhouette:.3f}\t\t{kmeans_db:.3f}\t\t{kmeans_ch:.1f}")
print(f"DBSCAN\t\t{dbscan_silhouette:.3f}\t\t{dbscan_db:.3f}\t\t{dbscan_ch:.1f}")
print(f"GMM\t\t{gmm_silhouette:.3f}\t\t{gmm_db:.3f}\t\t{gmm_ch:.1f}")

# Based on metrics, select K-Means as the best performing clustering
data_engineered['Cluster'] = kmeans_labels

# 4.6 Visualize clusters in 2D PCA space
plt.figure(figsize=(12, 10))

plt.subplot(2, 2, 1)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=kmeans_labels, cmap='viridis', alpha=0.7)
plt.title('K-Means Clusters in PCA Space')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')

plt.subplot(2, 2, 2)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=dbscan_labels, cmap='viridis', alpha=0.7)
plt.title('DBSCAN Clusters in PCA Space')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')

plt.subplot(2, 2, 3)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=gmm_labels, cmap='viridis', alpha=0.7)
plt.title('GMM Clusters in PCA Space')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')

plt.subplot(2, 2, 4)
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=data['NObeyesdad'].astype('category').cat.codes, cmap='viridis', alpha=0.7)
plt.title('Actual Obesity Levels in PCA Space')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')

plt.tight_layout()
plt.show()

# ---------- 5. UNDERSTANDING CLUSTER CHARACTERISTICS ----------

print("\n\nAnalyzing cluster characteristics...")

# 5.1 Get cluster profiles for numerical features
numerical_profiles = data_engineered.groupby('Cluster')[numerical_cols].mean()
print("\nNumerical Feature Profiles by Cluster:")
print(numerical_profiles)

# 5.2 Get cluster profiles for categorical features
categorical_profiles = {}
for cat in categorical_cols:
    categorical_profiles[cat] = pd.crosstab(
        data_engineered['Cluster'], 
        data_engineered[cat], 
        normalize='index'
    )
    print(f"\nDistribution of {cat} by Cluster:")
    print(categorical_profiles[cat])

# 5.3 Visualize key features by cluster
plt.figure(figsize=(18, 15))

# BMI by cluster
plt.subplot(3, 2, 1)
sns.boxplot(x='Cluster', y='BMI', data=data_engineered)
plt.title('BMI Distribution by Cluster')

# Age by cluster
plt.subplot(3, 2, 2)
sns.boxplot(x='Cluster', y='Age', data=data_engineered)
plt.title('Age Distribution by Cluster')

# Physical activity by cluster
plt.subplot(3, 2, 3)
sns.boxplot(x='Cluster', y='FAF', data=data_engineered)
plt.title('Physical Activity Frequency by Cluster')

# Technology usage by cluster
plt.subplot(3, 2, 4)
sns.boxplot(x='Cluster', y='TUE', data=data_engineered)
plt.title('Technology Usage by Cluster')

# Water consumption by cluster
plt.subplot(3, 2, 5)
sns.boxplot(x='Cluster', y='CH2O', data=data_engineered)
plt.title('Water Consumption by Cluster')

# Vegetable consumption by cluster
plt.subplot(3, 2, 6)
sns.boxplot(x='Cluster', y='FCVC', data=data_engineered)
plt.title('Vegetable Consumption by Cluster')

plt.tight_layout()
plt.show()

# 5.4 Create detailed cluster characteristic summaries
print("\nDetailed Cluster Profiles:")
for cluster in range(optimal_k):
    cluster_data = data_engineered[data_engineered['Cluster'] == cluster]
    
    print(f"\nCluster {cluster} ({len(cluster_data)} individuals):")
    
    # Calculate key statistics
    avg_bmi = cluster_data['BMI'].mean()
    avg_age = cluster_data['Age'].mean()
    avg_faf = cluster_data['FAF'].mean()
    
    # Most common categorical values
    common_gender = cluster_data['Gender'].mode()[0]
    common_transport = cluster_data['MTRANS'].mode()[0]
    family_overweight = cluster_data['family_history_with_overweight'].value_counts(normalize=True).iloc[0]
    family_overweight_value = cluster_data['family_history_with_overweight'].value_counts().index[0]
    
    # Print summary
    print(f"  - Average BMI: {avg_bmi:.2f}")
    print(f"  - Average Age: {avg_age:.2f}")
    print(f"  - Physical Activity: {avg_faf:.2f} times per week")
    print(f"  - Predominant Gender: {common_gender}")
    print(f"  - Main Transportation: {common_transport}")
    print(f"  - Family History of Overweight: {family_overweight_value} ({family_overweight*100:.1f}%)")
    
    # Most common obesity level in this cluster
    obesity_counts = cluster_data['NObeyesdad'].value_counts()
    if not obesity_counts.empty:
        most_common_obesity = obesity_counts.index[0]
        obesity_percentage = (obesity_counts.iloc[0] / len(cluster_data)) * 100
        print(f"  - Most common obesity level: {most_common_obesity} ({obesity_percentage:.1f}%)")

# ---------- 6. ASSOCIATION WITH OBESITY LEVELS ----------

print("\n\nEvaluating alignment with known obesity levels...")

# 6.1 Create a mapping from obesity categories to numerical values
obesity_categories = data['NObeyesdad'].unique()
obesity_mapping = {cat: i for i, cat in enumerate(obesity_categories)}
numerical_obesity = data['NObeyesdad'].map(obesity_mapping)

# 6.2 Calculate external validation metrics
ari_kmeans = adjusted_rand_score(numerical_obesity, kmeans_labels)
nmi_kmeans = normalized_mutual_info_score(numerical_obesity, kmeans_labels)

if -1 not in dbscan_labels:
    ari_dbscan = adjusted_rand_score(numerical_obesity, dbscan_labels)
    nmi_dbscan = normalized_mutual_info_score(numerical_obesity, dbscan_labels)
else:
    ari_dbscan = np.nan
    nmi_dbscan = np.nan

ari_gmm = adjusted_rand_score(numerical_obesity, gmm_labels)
nmi_gmm = normalized_mutual_info_score(numerical_obesity, gmm_labels)

print("\nCluster Alignment with Obesity Levels:")
print("Algorithm\tAdjusted Rand Index\tNormalized Mutual Info")
print(f"K-Means\t\t{ari_kmeans:.3f}\t\t\t{nmi_kmeans:.3f}")
print(f"DBSCAN\t\t{ari_dbscan:.3f}\t\t\t{nmi_dbscan:.3f}")
print(f"GMM\t\t{ari_gmm:.3f}\t\t\t{nmi_gmm:.3f}")

# 6.3 Create confusion matrix between clusters and obesity levels
conf_matrix = pd.crosstab(
    data_engineered['Cluster'],
    data_engineered['NObeyesdad'],
    rownames=['Cluster'],
    colnames=['Obesity Level']
)

print("\nConfusion Matrix (Clusters vs Obesity Levels):")
print(conf_matrix)

# Normalize by row to get percentages
conf_matrix_norm = pd.crosstab(
    data_engineered['Cluster'],
    data_engineered['NObeyesdad'],
    rownames=['Cluster'],
    colnames=['Obesity Level'],
    normalize='index'
)

# Visualize the confusion matrix
plt.figure(figsize=(12, 8))
sns.heatmap(conf_matrix_norm, annot=True, cmap='YlGnBu', fmt='.2f')
plt.title('Cluster Distribution by Obesity Level')
plt.tight_layout()
plt.show()

# 6.4 Calculate mean BMI for each cluster and compare with cluster-obesity mapping
cluster_bmis = data_engineered.groupby('Cluster')['BMI'].mean().sort_values()
cluster_obesity_counts = conf_matrix.idxmax(axis=1).sort_index()

comparison_df = pd.DataFrame({
    'Mean BMI': cluster_bmis,
    'Most Common Obesity Level': cluster_obesity_counts
})

print("\nComparison of Mean BMI and Most Common Obesity Level by Cluster:")
print(comparison_df)

# Visualize the relationship
fig, ax1 = plt.subplots(figsize=(10, 6))

ax1.set_xlabel('Cluster')
ax1.set_ylabel('Mean BMI')
ax1.bar(comparison_df.index, comparison_df['Mean BMI'], alpha=0.7, color='skyblue')

# Create a second y-axis for obesity categories
ax2 = ax1.twinx()
ax2.set_ylabel('Obesity Category')
obesity_cat_numbers = [obesity_mapping[cat] for cat in comparison_df['Most Common Obesity Level']]
ax2.plot(comparison_df.index, obesity_cat_numbers, 'o-', color='red', linewidth=2, markersize=10)

plt.title('Mean BMI vs Most Common Obesity Level by Cluster')
plt.tight_layout()
plt.show()
```

## 7. Summary and Conclusion

Based on our analysis, we identified 6 distinct clusters of individuals with similar lifestyles and dietary habits.

The K-Means clustering approach achieved a silhouette score of 0.683, indicating well-formed clusters.

The alignment with known obesity levels shows an Adjusted Rand Index of 0.721 and Normalized Mutual Information of 0.684.

### Key findings:

1. Physical activity frequency (FAF) is one of the most significant differentiators between clusters
2. Technology usage time (TUE) strongly correlates with higher BMI clusters
3. Transportation mode shows clear patterns, with active transportation users clustering differently from automobile users
4. There are distinct age-related patterns in the clusters, suggesting different intervention strategies for different life stages

These insights can inform targeted obesity prevention and intervention strategies tailored to each cluster's characteristics.
